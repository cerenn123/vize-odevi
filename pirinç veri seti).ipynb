```python
# 1. Kütüphaneleri Yükleme
import numpy as np
import pandas as pd
import requests # Yeni eklenen kütüphane
from io import StringIO # Yeni eklenen kütüphane

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# 2. Veri Setini İnternetten Yükleme (Pirinç Veri Seti)
print("Pirinç Veri Seti Yükleniyor...")

# Veri setini CSV formatında internetten doğrudan yükle
# URL hatası düzeltildi: 'main' yerine 'master' kullanıldı.
csv_url = 'https://raw.githubusercontent.com/erkansirin/ML-Datasets/master/Rice_Cammeo_and_Osmancik.csv'
try:
    response = requests.get(csv_url)
    response.raise_for_status()  # HTTP hatalarını yakala (örn. 404)
    df = pd.read_csv(StringIO(response.text))
    print("Veri Seti Başarıyla Yüklendi.")
except requests.exceptions.RequestException as e:
    print(f"Hata: Veri setini yüklerken bir sorun oluştu: {e}")
    df = pd.DataFrame() # Hata durumunda boş bir DataFrame ata
except Exception as e:
    # SyntaxError'u düzeltildi: f-string uygun şekilde kapatıldı.
    print(f"Hata: Beklenmeyen bir hata oluştu: {e}\n")
    df = pd.DataFrame() # Hata durumunda boş bir DataFrame ata

# df'in boş olup olmadığını kontrol et (eğer yükleme başarısız olduysa)
if df.empty:
    print("Hata: Veri seti yüklenemediği için sonraki adımlar atlanıyor.")
else:
    # İlk 5 satırı göster
    print("\nVeri Setinin İlk 5 Satırı (Örnek Özellikler):")
    print(df.head())

    # 3. Veri Ön İşleme ve Özellik Ayırma
    # Tüm sütunları X (giriş) olarak al, "Class" (sınıf) sütunu hariç
    X = df.drop('Class', axis=1)

    # "Class" (sınıf) sütununu y (çıktı) olarak al
    y_isim = df['Class']

    # Kategorik sınıf isimlerini sayısal değere çevir (0: Cammeo, 1: Osmancik)
    y = y_isim.astype('category').cat.codes

    print(f"\nKullanılan Özellikler (Pirinç Özellikleri): {list(X.columns)}")

    # 4. Veri Setini Eğitime ve Teste Bölme
    # Verinin %70'i eğitim, %30'u test için ayrılır
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    print(f"\nEğitim Veri Boyutu: {X_train.shape}")
    print(f"Test Veri Boyutu: {X_test.shape}")

    # 5. K-Nearest Neighbors (KNN) Modelini Eğitme
    # KNN modelini tanımla (k=5 komşu)
    knn_model = KNeighborsClassifier(n_neighbors=5)

    # Modeli eğitim verisi ile eğit
    knn_model.fit(X_train, y_train)

    # Test verisi üzerinde tahmin yap
    y_pred = knn_model.predict(X_test)

    # 6. Model Performansını Değerlendirme
    # Doğruluk Oranını (Accuracy Score) hesapla
    accuracy = accuracy_score(y_test, y_pred)

    # Sınıf isimlerini rapora eklemek için listeyi oluştur
    sinif_isimleri = y_isim.unique()

    # Detaylı sınıflandırma raporunu göster
    report = classification_report(y_test, y_pred, target_names=sinif_isimleri)

    print(f"\n*** SONUÇLAR ***")
    print(f"Model Doğruluk Oranı: {accuracy:.4f}")
    print("\nSınıflandırma Raporu:")
    print(report)
```
